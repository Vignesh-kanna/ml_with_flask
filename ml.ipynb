```python
import pandas as pd

# Sample DataFrame with missing values for Helmet Detection
data={
    'Age':[25, None, 30, 35, None, 4.5],
    'Height':[5.5, 6.1, 5.9, None, 5.8, 5.2],
    'HelmetStatus':[1, None, 1, 0, 1, 1],
    'Gender':['M', 'F', 'M', None, 'F', 'M'],
    'Experience':[2, 4, 6, 1, None, None]
}

df = pd.DataFrame(data)

print("Original DataFrame with Missing Values:")
print(df)

# Handling missing values by replacing with mean, median and mode

# Creating separate copies for each method of filling
df_filled_mean = df.copy()
df_filled_median = df.copy()
df_filled_mode = df.copy()

# Replacing missing values with the mean for numerical columns
df_filled_mean['Age'].fillna(df['Age'].mean(), inplace=True)
df_filled_mean['Height'].fillna(df['Height'].mean(), inplace=True)
df_filled_mean['Experience'].fillna(df['Experience'].mean(), inplace=True)

# Replacing missing values with the median for numerical columns for floating
df_filled_median['Age'].fillna(df['Age'].median(), inplace=True)
df_filled_median['Height'].fillna(df['Height'].median(), inplace=True)
df_filled_median['Experience'].fillna(df['Experience'].median(), inplace=True)

# Replacing missing values with the mode for categorical columns
df_filled_mode['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df_filled_mode['HelmetStatus'].fillna(df['HelmetStatus'].mode()[0], inplace=True)

# Show the results for all imputations
print("\nDataFrame after filling missing values with mean:")
print(df_filled_mean)
print("\nDataFrame after filling missing values with median:")
print(df_filled_median)
print("\nDataFrame after filling missing values with mode:")
print(df_filled_mode)
```

```python
import numpy as np
#Sample data (simple list of data points)
data = [10, 12, 14, 13, 100, 87, 5, 15, 16, 14, 13, 11]

#Step 1: Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

#Step 2: Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

#Step 3: Determine the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Step 4: Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

#Output the results
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR (Interquartile Range): {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers: {outliers}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Create a sample dataset with some outliers
np.random.seed(10)
data = pd.DataFrame({
     'value': np.concatenate([np.random.normal(0,1,100), np.random.normal(10, 1, 10)]) #Add outliers at the high end
 })

#Calular=te the IQR for box plot-based outlier detection
Q1 = data['value'].quantile(0.25)
Q3 = data['value'].quantile(0.75)
IQR = Q3-Q1

#Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Identify outliers
outliers = data[(data['value'] < lower_bound) | (data['value'] > upper_bound)]
#print outlier
print(f"Outliers based on Box plot criteria:\n{outliers}")

#Create a box plot and violin plot to visualize the daa and outliers
plt.figure(figsize=(12,6))

#Box plot
plt.subplot(1,2,1)
sns.boxplot(data=data)
plt.title('Box Plot')

#Violin plot with box plot inside for comparison
plt.subplot(1,2,2)
sns.violinplot(x=data['value'])
plt.title('Box Plot')

#Violin plot with box inside for comparison
plt.subplot(1, 2, 2)
sns.violinplot(x=data['value'])
plt.title('Violin plot')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = {
    "customer_id": [1, 2, 3, 4],
    "gender": ["Male","Female","Female","Male"],
    "subscription_status": ["Active","Inactive","Active","Inactive"]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

label_encoder = LabelEncoder()

df['gender_encoded'] = label_encoder.fit_transform(df['gender'])

df['subscription_status_encoded'] = label_encoder.fit_transform(df['subscription_status'])

print("\nDataFrame After Label Encoding:")
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Sample dataset
data = {
    "customer_id": [1, 2, 3, 4],
    "gender": ["Male", "Female", "Female", "Male"],
    "city": ["CityA", "CityB", "CityA", "CityC"]
}

# Create a DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Initialize OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse_output=False)

# Define columns to encode
columns_to_encode = ['gender', 'city']

# Fit and transform the data
encoded_data = one_hot_encoder.fit_transform(df[columns_to_encode])

# Get encoded column names
encoded_columns = one_hot_encoder.get_feature_names_out(columns_to_encode)

# Create a DataFrame for the encoded data
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns)

# Concatenate original DataFrame with encoded data
encoded_df_complete = pd.concat([df, encoded_df], axis=1)

print("\nDataFrame After One-Hot Encoding:")
print(encoded_df_complete)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Sample dataset
data = {
    "customer_id": [1, 2, 3, 4],
    "gender": ["Male", "Female", "Female", "Male"],
    "subscription_status": ["Active", "Inactive", "Active", "Inactive"],
    "city": ["CityA", "CityB", "CityA", "CityC"]
}

# Create a DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Initialize OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse_output=False)

# Define columns to encode
columns_to_encode = ['gender', 'city']

# Fit and transform the data
encoded_data = one_hot_encoder.fit_transform(df[columns_to_encode])

# Get encoded column names
encoded_columns = one_hot_encoder.get_feature_names_out(columns_to_encode)

# Create a DataFrame for the encoded data
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns)

# Concatenate original DataFrame with encoded data
encoded_df_complete = pd.concat([df, encoded_df], axis=1)

# Add ranking columns
encoded_df_complete['gender_rank'] = encoded_df_complete['gender'].rank(method='average').astype(int)
encoded_df_complete['subscription_status_rank'] = encoded_df_complete['subscription_status'].rank(method='average').astype(int)
encoded_df_complete['city_rank'] = encoded_df_complete['city'].rank(method='average').astype(int)

print("\nDataFrame After One-Hot Encoding and Ranking:")
print(encoded_df_complete)
```
